# programación orientada a objetos aplicada en la creación de gráficas

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

* Crear un lienzo
* En el lienzo se harán gráficas
* Las graficas se llaman *axes*
* La ventaja de usr el método orientado a objetos es que se pueden modificar las gráficas de forma individual y personalizar mejor la presentación de las gráficas



```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.2,0.55,0.4,0.3])

axes.plot(x,y,'b')
axes2.plot(y,x,'r')
axes2.set_facecolor('gray')

plt.show()
```

# Colores y estilos

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Commands
1. plt.style.available -> Retorna los estilos disponibles

```python
x = np.linspace(0.5,11)
```

```python
print(plt.style.available)
```

```python
plt.style.use('bmh')
```

```python
fig, ax = plt.subplots(figsize=(8,8))
```

```python
plt.style.use('classic')
fig, ax = plt.subplots(figsize=(8,8))
```

## Visualizar todos los estilos disponibles

```python
def mat_styles():
    styles = plt.style.available
    for style in styles:
        plt.style.use(style)
        fig, ax = plt.subplots(figsize=(8,8))
```

```python
mat_styles()
```

```python
plt.style.use('dark_background')
```

```python
fig, ax = plt.subplots(figsize=(8,8))

ax.plot(x,x+1,'r')
ax.plot(x,x+2,'y')
ax.plot(x,x+3,'g')
ax.plot(x,x+4,'w')
```

### Matlab

```python
fig, ax = plt.subplots(figsize=(8,8))

ax.plot(x,x+1,'r--')
ax.plot(x,x+2,'y-')
ax.plot(x,x+3,'g.-')
ax.plot(x,x+4,'w:')
```

## Pyplot

```python
fig, ax = plt.subplots(figsize=(8,8))

ax.plot(x,x+1, color = 'green')
ax.plot(x,x+2, color = 'blue')
ax.plot(x,x+3, color = '#fae7b5') # rgb
ax.plot(x,x+4, color = '#a2a2d0') # rgb
```

```python
fig, ax = plt.subplots(figsize=(8,8))

ax.plot(x,x+1, color = 'green', alpha=0.5) # alpha transparencia
ax.plot(x,x+2, color = 'blue', alpha=0.8, linewidth= 8) # linewidth grosor de línea
ax.plot(x,x+3, color = '#fae7b5',linestyle='--') # rgb, linestyle estilo de linea
ax.plot(x,x+4, color = '#a2a2d0', linestyle='dashed') # rgb
```

```python
fig, ax = plt.subplots(figsize=(8,8))

ax.plot(x,x+1, color = 'green', alpha=0.5, marker='o',markersize=5,
       markerfacecolor= '#fe6f5e') # alpha transparencia
ax.plot(x,x+2, color = 'blue', alpha=0.8, linewidth= 2, marker='x') # linewidth grosor de línea
ax.plot(x,x+3, color = '#fae7b5',linestyle='--', marker='v') # rgb, linestyle estilo de linea
ax.plot(x,x+4, color = '#a2a2d0', linestyle='dashed', marker= 'P') # rgb
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

# Set
* Es un comando que permite cambiar colores, estilos de las gr√°ficas en seaborn

```python
sns.barplot(x=['A','B','C'],y=[1,3,2])
plt.show()
```

```python
sns.set(style='dark')
sns.barplot(x=['A','B','C'],y=[1,3,2])
plt.show()
```

```python
sns.set(style='dark', palette='dark')# deep, muted
sns.barplot(x=['A','B','C'],y=[1,3,2])
plt.show()
```

```python
sns.set(style='dark', palette='muted',font="Verdana",font_scale=1)
sns.barplot(x=['A','B','C'],y=[1,3,2])
plt.show()
```

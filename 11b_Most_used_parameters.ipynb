```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# datos precargados en seaborn
tips = sns.load_dataset('tips')
```

## Comandos

* displot -> graficas de distribución
* kind= -> qué tipo de gráfico se quiere usar

```python
tips
```

```python
sns.displot(data=tips,x='total_bill')
plt.show()
```

```python
sns.displot(data=tips,x='total_bill', y='tip')
plt.show()
```

```python
sns.displot(data=tips,x='total_bill', y='tip', hue='sex')
plt.show()
```

```python
sns.displot(data=tips,x='total_bill', hue='sex') # hue, agrupa por 'sex' solo x='tip'
plt.show()
```

```python
sns.displot(data=tips,x='total_bill', hue='sex', kind='kde') # kind, tipo de gráfico a usar
plt.show()
```

```python
sns.displot(data=tips,x='total_bill', hue='sex', kind='kde', legend=False) # 
plt.show()
```

```python
sns.displot(data=tips,x='total_bill', hue='sex', kind='kde', legend=False, palette='dark') # 
plt.show()
```

```python
sns.displot(data=tips,x='total_bill', hue='sex', kind='kde', legend=False, palette='dark', alpha=0.5) # 
plt.show()
```

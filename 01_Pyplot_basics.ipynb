# Creación de gráficas
1. Histogramas
2. Pie chart
3. Scatter
4. Boxplot
5. Personalizar en las gráficas

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Crear array
x = np.linspace(0,5,11)
```

```python
# Crear y
y = x ** 2
```

```python
# Crear la primera gráfica
plt.plot(x,y)
plt.show()
```

```python
# modificar gráfica
```

```python
plt.plot(x,y,'r') # pintar la líne de color 
plt.show()
```

```python
plt.plot(x,y,'y') # pintar la líne de color 
plt.show()
```

```python
plt.plot(x,y,'ro') # agregar markers tipo o
plt.show()
```

```python
plt.plot(x,y,'rx') # agregar markers tipo x
plt.show()
```

```python
plt.plot(x,y,'rs-') # agregar markers tipo squads con continuidad en el marker, osease una linea
plt.show()
```

```python
plt.plot(x,y,'bD:') # agregar markers tipo diamonds con línea punteada 
plt.show()
```

```python
plt.hist(x)
plt.show()
```

```python
plt.pie(x)
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.boxplot(x) # distribución de puntos
plt.show()
```

# Títulos, leyendas, etiquetas, tamaño

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Comandos

### Título
* set_title()
* plt.title('')

### Tamaño
* figsize=()

### Etiquetas
* set_xlabel()
* set_ylabel()
* label = ()
* plt.xlabel('')
* plt.ylabel('')

### Leyendas
* legend()
* plt.legend(loc='lower left)  **investigar mas**

### Notación Matemática
* $ ecuacion $

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots(1,2, figsize=(10,5))
axes[0].plot(x,y, label="sin(x)")
axes[0].set_title('Relacion X-Y')
axes[0].set_xlabel('exis')
axes[0].set_ylabel('ye')
axes[0].legend()
axes[1].plot(y,x)
axes[1].set_title('Relacion Y-X')
axes[1].set_xlabel('ye')
axes[1].set_ylabel('exis')
plt.show()
```

```python
fig, axes = plt.subplots(1,2, figsize=(10,5))
axes[0].plot(x,y, label="$sin(x)$") # con notación matemática
axes[0].set_title('Relacion X-Y')
axes[0].set_xlabel('exis')
axes[0].set_ylabel('ye')
axes[0].legend()
axes[1].plot(y,x)
axes[1].set_title('Relacion Y-X')
axes[1].set_xlabel('ye')
axes[1].set_ylabel('exis')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots(1,2, figsize=(10,5))
axes[0].plot(x,y)
axes[0].set_title('Relacion X-Y')
axes[0].set_xlabel('exis')
axes[0].set_ylabel('ye')
axes[1].plot(y,x)
axes[1].set_title('Relacion Y-X')
axes[1].set_xlabel('ye')
axes[1].set_ylabel('exis')
plt.show()
```

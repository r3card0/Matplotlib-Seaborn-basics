```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1,50,100) # del 1 al 50, 100 ejemplos
```

## Analizar distribuci√≥n y frecuencia. Histograma

```python
plt.hist(data)
plt.show()
```

```python
plt.hist(data,bins=50) # por defualt el histograma muestra 10 bins
plt.show()
```

```python
plt.hist(data,bins=50,width=5)
plt.show()
```

```python
plt.hist(data,bins=10,histtype='step')
plt.show()
```

## Boxplot

```python
plt.boxplot(data)
plt.show()
```

```python
plt.boxplot(data,vert=False)
plt.show()
```

```python
plt.boxplot(data,vert=False,patch_artist=True)
plt.show()
```

```python
plt.boxplot(data,vert=False,patch_artist=True,notch=True)
plt.show()
```

```python
data = np.append(data,200)
plt.boxplot(data,vert=False,patch_artist=True,notch=True)
plt.show()
```

```python
data = np.append(data,200)
plt.boxplot(data,vert=False,patch_artist=True,notch=True,showfliers=False)
plt.show()
```

## Grafico de dispersi√≥n. Scatter

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * x ) ** 2
colors = np.random.rand(N)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y,s=area)
plt.show()
```

```python
plt.scatter(x,y,s=area,c=colors)
plt.show()
```

```python
plt.scatter(x,y,s=area,c=colors,marker='p')
plt.show()
```

```python
plt.scatter(x,y,s=area,c=colors,marker='p',alpha=0.5)
plt.show()
```
